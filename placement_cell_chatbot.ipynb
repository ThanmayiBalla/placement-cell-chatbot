```python
print("=== Placement Cell ChatBot ===")

name = input("Enter your name: ")

print(f"\nWelcome {name}!")
print("Ask me about:")
print("companies, eligibility, training, internship, package, contact")
print("Type 'bye' to exit.\n")

while True:

    user = input("You: ").lower()

    if "companies" in user:
        print("Bot: TCS, Infosys, Wipro, Accenture and Cognizant visit our campus.")

    elif "eligibility" in user:
        print("Bot: Students must have at least 60% in academics with no active backlogs.")

    elif "training" in user:
        print("Bot: Aptitude, Coding and Soft Skills training are provided.")

    elif "internship" in user:
        print("Bot: Internship opportunities are available through college partnerships.")

    elif "package" in user:
        print("Bot: Average package is 4.5 LPA and highest package is 12 LPA.")

    elif "contact" in user:
        print("Bot: Placement Cell Contact Number: 9876543210")

    elif "help" in user:
        print("Bot: You can ask about companies, eligibility, training, internship, package or contact.")

    elif user == "bye":
        print(f"Bot: Thank you {name}. Best of luck for your placements!")
        break

    else:
        print("Bot: Sorry, I don't understand. Type 'help' for available options.")
```
